```python
import numpy as np
import matplotlib.pyplot as plt
gamma1 = 1.0
gamma2 = 1.5
gamma3 = 1.8
gamma4 = 2.5 # Damping Gamma
w_0 = 105
w_star1 = np.sqrt(((w_0)**2) - (gamma1**2))
w_star2 = np.sqrt(((w_0)**2) - (gamma2**2))
w_star3 = np.sqrt(((w_0)**2) - (gamma3**2))
w_star4 = np.sqrt(((w_0)**2) - (gamma4**2))
a = 0.004
```

```python
t = np.arange(0, 2, 0.01).tolist()
v_t1 = []
v_t2 = []
v_t3 = []
v_t4 = []
for i in t:
    v_t1.append((a*np.exp(-gamma1*i))*((gamma1*np.cos((w_star1*i)+(np.pi/4)))+(w_star1*np.sin((w_star1*i)+(np.pi/4)))))

for i in t:
    v_t2.append((a*np.exp(-gamma2*i))*((gamma2*np.cos((w_star2*i)+(np.pi/4)))+(w_star2*np.sin((w_star2*i)+(np.pi/4)))))
    
for i in t:
    v_t3.append((a*np.exp(-gamma3*i))*((gamma3*np.cos((w_star3*i)+(np.pi/4)))+(w_star3*np.sin((w_star3*i)+(np.pi/4)))))
    
for i in t:
    v_t4.append((a*np.exp(-gamma4*i))*((gamma4*np.cos((w_star4*i)+(np.pi/4)))+(w_star4*np.sin((w_star4*i)+(np.pi/4)))))
    
```
